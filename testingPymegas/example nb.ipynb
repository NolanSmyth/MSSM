```python
import pandas as pd

import os
import sys

sys.path.append("/media/jt/data/Projects/MSSM/micromegas_5.2.7.a/MSSM/")
from pymicromegas import MicromegasSettings, SugraParameters
from softsusy import softsusy
from spheno import spheno
from suspect import suspect

data_dir = "/media/jt/data/Projects/MSSM/data/Hollingsworth"
sys.path.append(data_dir)
from read_dataset import import_data
```

```python
dataset = "cMSSM"
df = import_data(os.path.join(data_dir, dataset), dataset.lower())
df = df[df['sgnmu'] == 1]
sugras = df[["m0", "m12", "a0", "tanb", "sgnmu"]]
sugras.head()
```

```python
settings = MicromegasSettings(
    relic_density=True,
    masses=True,
    gmuon=True,
    bsg=True,
    bsmumu=True,
    btaunu=True,
    delta_rho=True,
    sort_odd=True,
    fast=False,
    beps=0.0001,
    cut=0.01,
)
```

```python
# Run with default settings
batch = sugras.head(3).apply(lambda x: SugraParameters(*x), axis=1).to_list()

softsusy_results = softsusy(batch, settings=settings)
spheno_results = spheno(batch, settings=settings)
suspect_results = suspect(batch, settings=settings)
```

```python
def results_to_df(res):
    columns = [
        "omega",
        "xf",
        "bsgsm",
        "bsgnlo",
        "deltarho",
        "bsmumu",
        "btaunu",
        "gmuon",
        "msne",
        "msnm",
        "msel",
        "mser",
        "msml",
        "msmr",
        "msdl",
        "msdr",
        "msul",
        "msur",
        "mssl",
        "mssr",
        "mscl",
        "mscr",
        "msnl",
        "msl1",
        "msl2",
        "msb1",
        "msb2",
        "mst1",
        "mst2",
        "mg",
        "mneut1",
        "mneut2",
        "mneut3",
        "mneut4",
        "mchg1",
        "mchg2",
        "mhsm",
        "mh",
        "mhc",
    ]
    res_dict = {}
    for key in columns:
        values = getattr(res, key)()
        repeated_val_idx = len(values) // len(res.omega())
        res_dict[key] = values[::repeated_val_idx]

    res_df = pd.DataFrame(res_dict)
    return res_df
```
